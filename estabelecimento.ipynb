##### üü° Em uma analise por data type √© preciso ajustar ' COD ATIVIDADE PRINCIPAL'; 'CEP' ; 'CNPJ BASICO', 'CNPJ ORDEM', 'CNPJ DV' para object pois eles n√£o possuem a fun√ß√£o de soma visto que se trata de c√≥digos
##### üü° Em uma analise por data type √© preciso ajustar 'DATA DA SITUA√á√ÉO CADASTRAL' ; 'DATA DE IN√çCIO DE ATIVIDADE' ; 'DATA DA SITUA√á√ÉO ESPECIAL' para o formato de data

##### üü©üü© Alterando Dtype para object 

```python
conversao = ['COD ATIVIDADE PRINCIPAL', 'CEP' , 'CNPJ BASICO', 'CNPJ ORDEM', 'CNPJ DV']

# Converter as colunas espec√≠ficas de 'int' para 'object'
df[conversao] = df[conversao].astype('object')

# Verificar os tipos de dados ap√≥s a convers√£o
df.dtypes
```

```python
df['CNPJ BASICO'] = df['CNPJ BASICO'].astype(str)

# Adicionar zeros √† esquerda para garantir 8 caracteres
df['CNPJ BASICO'] = df['CNPJ BASICO'].str.zfill(8)
```

```python
df = df.sort_values(by='CNPJ BASICO', ascending=True)

df
```

```python
# Definir as colunas espec√≠ficas para verificar
colunas_data = ['DATA DA SITUA√á√ÉO CADASTRAL','DATA DE IN√çCIO DE ATIVIDADE','DATA DA SITUA√á√ÉO ESPECIAL']
df[colunas_data]
```

```python
valores_unicos = df['DATA DA SITUA√á√ÉO ESPECIAL'].unique()
print(valores_unicos)
```

```python
# Substituir valores '0' ou inv√°lidos por NaT (Not a Time)
df['DATA DA SITUA√á√ÉO CADASTRAL'] = df['DATA DA SITUA√á√ÉO CADASTRAL'].replace('0', pd.NaT)
df['DATA DE IN√çCIO DE ATIVIDADE'] = df['DATA DE IN√çCIO DE ATIVIDADE'].replace('0', pd.NaT)
df['DATA DA SITUA√á√ÉO ESPECIAL'] = df['DATA DA SITUA√á√ÉO ESPECIAL'].replace('0', pd.NaT)

# Converter para o formato datetime, for√ßando o tratamento de valores inv√°lidos
df['DATA DA SITUA√á√ÉO CADASTRAL'] = pd.to_datetime(df['DATA DA SITUA√á√ÉO CADASTRAL'], errors='coerce', format='%Y%m%d')
df['DATA DE IN√çCIO DE ATIVIDADE'] = pd.to_datetime(df['DATA DE IN√çCIO DE ATIVIDADE'], errors='coerce', format='%Y%m%d')
df['DATA DA SITUA√á√ÉO ESPECIAL'] = pd.to_datetime(df['DATA DA SITUA√á√ÉO ESPECIAL'], errors='coerce', format='%Y%m%d')

# Formatar as datas para DD/MM/AAAA
df['DATA DA SITUA√á√ÉO CADASTRAL'] = df['DATA DA SITUA√á√ÉO CADASTRAL'].dt.strftime('%d/%m/%Y')
df['DATA DE IN√çCIO DE ATIVIDADE'] = df['DATA DE IN√çCIO DE ATIVIDADE'].dt.strftime('%d/%m/%Y')
df['DATA DA SITUA√á√ÉO ESPECIAL'] = df['DATA DA SITUA√á√ÉO ESPECIAL'].dt.strftime('%d/%m/%Y')

# Exibir as colunas para verificar a convers√£o
df[['DATA DA SITUA√á√ÉO CADASTRAL', 'DATA DE IN√çCIO DE ATIVIDADE', 'DATA DA SITUA√á√ÉO ESPECIAL']]
```

##### Analises:

* QUANTIDADE DE CNPJ POR BAIRRO 
* QUANTIDADE DE CNPJ POR SITUA√á√ÉO CADASTRAL
* TIPO DE LOGRADOURO
* CNPJ POR PA√çS
* QUANTIDADE DE MATRIZ E FILIAL

```python
cnpj_por_bairro = (
    df.groupby('BAIRRO')['CNPJ BASICO']
    .nunique()  # Conta valores √∫nicos
    .sort_values(ascending=False)  # Ordena em ordem decrescente
)

cnpj_por_bairro
```

```python
# Lista de bairros para an√°lise
bairros = ['BOA VIAGEM', 'BOA VISTA', 'IMBIRIBEIRA', 'PINA', 'CASA AMARELA']

# Resultado: tipo de logradouro mais frequente por bairro
for bairro in bairros:
    subset = df[df['BAIRRO'] == bairro]  # Filtrar pelo bairro
    tipo_mais_frequente = subset['TIPO DO LOGRADOURO'].mode()[0]  # Moda do tipo de logradouro
    print(f"Bairro: {bairro} - Tipo de logradouro mais frequente: {tipo_mais_frequente}")
```

```python
cnpj_por_logradouro = (
    df.groupby('TIPO DO LOGRADOURO')['CNPJ BASICO']
    .nunique()  # Conta os valores √∫nicos de CNPJs
    .sort_values(ascending=False)  # Ordena em ordem decrescente
)
cnpj_por_logradouro
```

```python
cnpj_por_pais = (
    df.groupby('PAIS')['CNPJ BASICO']
    .nunique()  # Conta os valores √∫nicos de CNPJs
    .sort_values(ascending=False)  # Ordena em ordem decrescente
)

cnpj_por_pais
```

```python
# Contar a quantidade de cada valor na coluna 'MATRIZ/FILIAL'
counts = df['MATRIZ/FILIAL'].value_counts()

# Quantidade de 'MATRIZ' e 'FILIAL'
contagem_matriz = counts.get('MATRIZ', 0)
contagem_filial = counts.get('FILIAL', 0)

print(f"Qnt 'MATRIZ': {contagem_matriz}")
print(f"Qnt 'FILIAL': {contagem_filial}")
```

```python
cnpj_por_situacao = (
    df.groupby('SITUA√á√ÉO CADASTRAL')['CNPJ BASICO']
    .nunique()  # Conta os valores √∫nicos de CNPJs
    .sort_values(ascending=False)  # Ordena em ordem decrescente
)

cnpj_por_situacao
```
