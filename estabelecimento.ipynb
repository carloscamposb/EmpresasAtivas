##### 🟩🟩 Importando Bibliotecas

```python
import pandas as pd
import numpy as np
import os
```

```python
# Diretório onde os arquivos CSV estão localizados
CSV_FOLDER = r'./datasets/CSV_Separados'

# Nome do arquivo CSV que você quer carregar
CSV_FILE = 'ESTABELECIMENTOS.csv'  # Substitua pelo nome do arquivo gerado

# Caminho completo do arquivo CSV
csv_path = os.path.join(CSV_FOLDER, CSV_FILE)

# Verificar se o arquivo existe antes de tentar carregá-lo
if os.path.exists(csv_path):
    # Carregar o arquivo CSV como DataFrame
    df = pd.read_csv(csv_path)
    print(f"Arquivo CSV '{CSV_FILE}' carregado com sucesso!")
else:
    print(f"Erro: O arquivo '{csv_path}' não foi encontrado.")

pd.set_option('display.max_columns', None)  # Exibe todas as colunas
pd.set_option('display.max_colwidth', None)  # Exibe o conteúdo completo de cada célula
pd.set_option('display.width', None)  # Ajuste de largura para evitar quebra de linha
```

##### 🟡 Em uma analise por data type é preciso ajustar ' COD ATIVIDADE PRINCIPAL'; 'CEP' ; 'CNPJ BASICO', 'CNPJ ORDEM', 'CNPJ DV' para object pois eles não possuem a função de soma visto que se trata de códigos
##### 🟡 Em uma analise por data type é preciso ajustar 'DATA DA SITUAÇÃO CADASTRAL' ; 'DATA DE INÍCIO DE ATIVIDADE' ; 'DATA DA SITUAÇÃO ESPECIAL' para o formato de data

##### 🟩🟩 Alterando Dtype para object 

```python
conversao = ['COD ATIVIDADE PRINCIPAL', 'CEP' , 'CNPJ BASICO', 'CNPJ ORDEM', 'CNPJ DV']

# Converter as colunas específicas de 'int' para 'object'
df[conversao] = df[conversao].astype('object')

# Verificar os tipos de dados após a conversão
df.dtypes
```

```python
df['CNPJ BASICO'] = df['CNPJ BASICO'].astype(str)

# Adicionar zeros à esquerda para garantir 8 caracteres
df['CNPJ BASICO'] = df['CNPJ BASICO'].str.zfill(8)
```

```python
df = df.sort_values(by='CNPJ BASICO', ascending=True)

df
```

```python
# Definir as colunas específicas para verificar
colunas_data = ['DATA DA SITUAÇÃO CADASTRAL','DATA DE INÍCIO DE ATIVIDADE','DATA DA SITUAÇÃO ESPECIAL']
df[colunas_data]
```

```python
valores_unicos = df['DATA DA SITUAÇÃO ESPECIAL'].unique()
print(valores_unicos)
```

```python
# Substituir valores '0' ou inválidos por NaT (Not a Time)
df['DATA DA SITUAÇÃO CADASTRAL'] = df['DATA DA SITUAÇÃO CADASTRAL'].replace('0', pd.NaT)
df['DATA DE INÍCIO DE ATIVIDADE'] = df['DATA DE INÍCIO DE ATIVIDADE'].replace('0', pd.NaT)
df['DATA DA SITUAÇÃO ESPECIAL'] = df['DATA DA SITUAÇÃO ESPECIAL'].replace('0', pd.NaT)

# Converter para o formato datetime, forçando o tratamento de valores inválidos
df['DATA DA SITUAÇÃO CADASTRAL'] = pd.to_datetime(df['DATA DA SITUAÇÃO CADASTRAL'], errors='coerce', format='%Y%m%d')
df['DATA DE INÍCIO DE ATIVIDADE'] = pd.to_datetime(df['DATA DE INÍCIO DE ATIVIDADE'], errors='coerce', format='%Y%m%d')
df['DATA DA SITUAÇÃO ESPECIAL'] = pd.to_datetime(df['DATA DA SITUAÇÃO ESPECIAL'], errors='coerce', format='%Y%m%d')

# Formatar as datas para DD/MM/AAAA
df['DATA DA SITUAÇÃO CADASTRAL'] = df['DATA DA SITUAÇÃO CADASTRAL'].dt.strftime('%d/%m/%Y')
df['DATA DE INÍCIO DE ATIVIDADE'] = df['DATA DE INÍCIO DE ATIVIDADE'].dt.strftime('%d/%m/%Y')
df['DATA DA SITUAÇÃO ESPECIAL'] = df['DATA DA SITUAÇÃO ESPECIAL'].dt.strftime('%d/%m/%Y')

# Exibir as colunas para verificar a conversão
df[['DATA DA SITUAÇÃO CADASTRAL', 'DATA DE INÍCIO DE ATIVIDADE', 'DATA DA SITUAÇÃO ESPECIAL']]
```

```python
valores_unicos = df['DATA DA SITUAÇÃO ESPECIAL'].unique()
print(valores_unicos)
```

##### Analises:

* QUANTIDADE DE CNPJ POR BAIRRO 
* QUANTIDADE DE CNPJ POR SITUAÇÃO CADASTRAL
* TIPO DE LOGRADOURO
* CNPJ POR PAÍS
* QUANTIDADE DE MATRIZ E FILIAL

```python
cnpj_por_bairro = (
    df.groupby('BAIRRO')['CNPJ BASICO']
    .nunique()  # Conta valores únicos
    .sort_values(ascending=False)  # Ordena em ordem decrescente
)

cnpj_por_bairro
```

```python
# Lista de bairros para análise
bairros = ['BOA VIAGEM', 'BOA VISTA', 'IMBIRIBEIRA', 'PINA', 'CASA AMARELA']

# Resultado: tipo de logradouro mais frequente por bairro
for bairro in bairros:
    subset = df[df['BAIRRO'] == bairro]  # Filtrar pelo bairro
    tipo_mais_frequente = subset['TIPO DO LOGRADOURO'].mode()[0]  # Moda do tipo de logradouro
    print(f"Bairro: {bairro} - Tipo de logradouro mais frequente: {tipo_mais_frequente}")
```

```python
cnpj_por_logradouro = (
    df.groupby('TIPO DO LOGRADOURO')['CNPJ BASICO']
    .nunique()  # Conta os valores únicos de CNPJs
    .sort_values(ascending=False)  # Ordena em ordem decrescente
)
cnpj_por_logradouro
```

```python
cnpj_por_pais = (
    df.groupby('PAIS')['CNPJ BASICO']
    .nunique()  # Conta os valores únicos de CNPJs
    .sort_values(ascending=False)  # Ordena em ordem decrescente
)

cnpj_por_pais
```

```python
# Contar a quantidade de cada valor na coluna 'MATRIZ/FILIAL'
counts = df['MATRIZ/FILIAL'].value_counts()

# Quantidade de 'MATRIZ' e 'FILIAL'
contagem_matriz = counts.get('MATRIZ', 0)
contagem_filial = counts.get('FILIAL', 0)

print(f"Qnt 'MATRIZ': {contagem_matriz}")
print(f"Qnt 'FILIAL': {contagem_filial}")
```

```python
cnpj_por_situacao = (
    df.groupby('SITUAÇÃO CADASTRAL')['CNPJ BASICO']
    .nunique()  # Conta os valores únicos de CNPJs
    .sort_values(ascending=False)  # Ordena em ordem decrescente
)

cnpj_por_situacao
```
