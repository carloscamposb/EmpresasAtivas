##### 游릴游릴 Dividindo o arquivo EmpresasAtivas.xlsx em m칰ltiplos arquivos csv

```python
import pandas as pd
import os

# Caminho do arquivo Excel com v치rias abas
INPUT_PATH = r'C:\Users\carli\Desktop\EmpresasAtivas\datasets\EmpresasAtivas.xlsx'

# Pasta para salvar os arquivos CSV
OUTPUT_FOLDER = r'C:\Users\carli\Desktop\EmpresasAtivas\datasets\CSV_Separados'

# Verificando se o arquivo Excel existe
if not os.path.exists(INPUT_PATH):
    print(f"Erro: O arquivo Excel '{INPUT_PATH}' n칚o foi encontrado.")
    exit()

# Criando a pasta de sa칤da, se n칚o existir
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

try:
    # Lendo todas as abas do arquivo Excel
    print("Carregando o arquivo Excel...")
    excel_data = pd.read_excel(INPUT_PATH, sheet_name=None)  # `sheet_name=None` carrega todas as abas
    print(f"Arquivo carregado com sucesso! Abas encontradas: {list(excel_data.keys())}")
except Exception as e:
    print(f"Erro ao carregar o arquivo Excel: {e}")
    exit()

# Salvando cada aba como um arquivo CSV separado
print("Iniciando a separa칞칚o das abas em arquivos CSV...")
for sheet_name, data in excel_data.items():
    try:
        # Nome do arquivo CSV baseado no nome da aba
        output_file = os.path.join(OUTPUT_FOLDER, f"{sheet_name}.csv")
        
        # Salvando a aba como CSV
        data.to_csv(output_file, index=False, encoding='utf-8')
        print(f"Aba '{sheet_name}' salva como CSV: {output_file}")
    except Exception as e:
        print(f"Erro ao salvar a aba '{sheet_name}' como CSV: {e}")

print("Processo conclu칤do! Todas as abas foram processadas.")

```

```python
# Substituir valores '0' ou inv치lidos por NaT (Not a Time)
dados['DATA DE ADESAO AO SIMPLES'] = dados['DATA DE ADESAO AO SIMPLES'].replace('0', pd.NaT)
dados['DATA DE EXCLUSAO DO SIMPLES'] = dados['DATA DE EXCLUSAO DO SIMPLES'].replace('0', pd.NaT)
dados['DATA DE ADESAO AO MEI'] = dados['DATA DE ADESAO AO MEI'].replace('0', pd.NaT)
dados['DATA DE EXCLUSAO DO MEI'] = dados['DATA DE EXCLUSAO DO MEI'].replace('0', pd.NaT)

# Converter para o formato datetime, for칞ando o tratamento de valores inv치lidos
dados['DATA DE ADESAO AO SIMPLES'] = pd.to_datetime(dados['DATA DE ADESAO AO SIMPLES'], errors='coerce', format='%Y%m%d')
dados['DATA DE EXCLUSAO DO SIMPLES'] = pd.to_datetime(dados['DATA DE EXCLUSAO DO SIMPLES'], errors='coerce', format='%Y%m%d')
dados['DATA DE ADESAO AO MEI'] = pd.to_datetime(dados['DATA DE ADESAO AO MEI'], errors='coerce', format='%Y%m%d')
dados['DATA DE EXCLUSAO DO MEI'] = pd.to_datetime(dados['DATA DE EXCLUSAO DO MEI'], errors='coerce', format='%Y%m%d')

# Formatar as datas para DD/MM/AAAA
dados['DATA DE ADESAO AO SIMPLES'] = dados['DATA DE ADESAO AO SIMPLES'].dt.strftime('%d/%m/%Y')
dados['DATA DE EXCLUSAO DO SIMPLES'] = dados['DATA DE EXCLUSAO DO SIMPLES'].dt.strftime('%d/%m/%Y')
dados['DATA DE ADESAO AO MEI'] = dados['DATA DE ADESAO AO MEI'].dt.strftime('%d/%m/%Y')
dados['DATA DE EXCLUSAO DO MEI'] = dados['DATA DE EXCLUSAO DO MEI'].dt.strftime('%d/%m/%Y')

# Exibir as colunas para verificar a convers칚o
dados[['DATA DE ADESAO AO SIMPLES', 'DATA DE EXCLUSAO DO SIMPLES', 'DATA DE ADESAO AO MEI','DATA DE EXCLUSAO DO MEI']]
```
