```python
# Carregando os dados do CSV
df_empresas = pd.read_csv('C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\empresas.csv', sep=';', decimal=',', encoding='UTF-8')
df_estabelecimento = pd.read_csv('C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\estabelecimento.csv', sep=';', decimal=',', encoding='UTF-8')
df_simples = pd.read_csv('C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\simples.csv', sep=';', decimal=',', encoding='UTF-8')
df_socio = pd.read_csv('C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\socios.csv', sep=';', decimal=',', encoding='UTF-8')
df_cnae = pd.read_csv('C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\cnae.csv', sep=';', decimal=',', encoding='UTF-8')
```

```python
df_estabelecimento= df_estabelecimento.rename(columns={'COD ATIVIDADE PRINCIPAL': 'CODIGO CNAE'})
```

```python
df_cnae= df_cnae.rename(columns={'CODIGO': 'CODIGO CNAE'})
```

```python
merge_df = df_empresas.merge(df_socio, on='CNPJ BASICO', how='left')  
merge_df = merge_df.merge(df_simples, on='CNPJ BASICO', how='left')
merge_df = merge_df.merge(df_estabelecimento, on='CNPJ BASICO', how='left')
merge_df = merge_df.merge(df_cnae, on='CODIGO CNAE', how='left')


merge_df
```

```python
# Selecionar apenas as colunas D1, D2 e D3
df_empresasAtivas = merge_df[['CNPJ BASICO', 'CNPJ ORDEM', 'CNPJ DV', 'MATRIZ/FILIAL','PAIS','SITUAÇÃO CADASTRAL', 'DATA DA SITUAÇÃO CADASTRAL','DATA DE INÍCIO DE ATIVIDADE','ATIVIDADE PRINCIPAL','CODIGO CNAE','CEP']]

df_empresasAtivas
```

```python
df_empresasAtivas['CNPJ BASICO'] = df_empresasAtivas['CNPJ BASICO'].astype(str)

# Adicionar zeros à esquerda para garantir 8 caracteres
df_empresasAtivas['CNPJ BASICO'] = df_empresasAtivas['CNPJ BASICO'].str.zfill(8)
```
