```python
import pandas as pd
import requests
```

```python
df= pd.read_csv('C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\faltantes_bairro.csv', sep=';', decimal=',', encoding='UTF-8')
```

```python
df.drop_duplicates()
```

```python
# Caminho do arquivo CSV de entrada
arquivo_csv = 'C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados'

# Número de linhas por arquivo dividido
tamanho_lote = 1000

# Dividir o DataFrame em partes de 1000 linhas
for i in range(0, len(df), tamanho_lote):
    # Criar uma fatia do DataFrame
    df_parte = df.iloc[i:i + tamanho_lote]
    
    # Salvar cada parte em um novo arquivo CSV
    nome_arquivo = f'parte_{i // tamanho_lote + 1}.csv'
    df_parte.to_csv(f'C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\{nome_arquivo}', index=False, sep=';')
    print(f'Arquivo {nome_arquivo} salvo com sucesso.')
```

```python
# Função para buscar o bairro a partir do CEP
def encontrar_bairro(cep, tentativas=1):
    # Verificar se o CEP é 'Não Informado' ou 0
    if cep == 'Não Informado' or cep == 0:
        return 'Não Informado'  # Ou np.nan se preferir um valor nulo
    
    url = f'https://opencep.com/v1/{cep}.json'
    for tentativa in range(tentativas):
        try:
            resposta = requests.get(url, timeout=5)  # Timeout de 30 segundos
            resposta.raise_for_status()  # Levanta uma exceção se o código de status não for 200
            dados = resposta.json()

            # Verifica se a resposta tem erro ou dados faltantes
            if 'erro' in dados:
                return 'Não informado'
            
            # Retorna o bairro encontrado
            return dados.get('bairro', 'Não informado')
        
        except requests.exceptions.RequestException as e:
            print(f"Erro ao buscar o CEP {cep}: {e}")
            if tentativa < tentativas - 1:
                print(f"Tentando novamente... ({tentativa + 1}/{tentativas})")
            else:
                return 'Não informado'
    
    return 'Não informado'

# Função principal para atualizar os bairros de uma vez
def atualizar_bairros(busca):
    # Iterar explicitamente em ordem de ID
    for index, row in busca.iterrows():
        cep = row['CEP']
        bairro_atualizado = encontrar_bairro(cep)
        busca.at[index, 'BAIRRO_Novo'] = bairro_atualizado

    return busca

# Ler o arquivo faltantes_bairro.csv
parte_8=pd.read_csv('C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\parte_8.csv', sep=';') 

# Atualizar a coluna BAIRRO_Novo
parte_8 = atualizar_bairros(parte_8)

# Salvar o resultado atualizado em um novo arquivo
parte_8.to_csv('C:\\Users\\carli\\Desktop\\EmpresasAtivas\\datasets\\CSV_Tratados\\bairro_atualizado_8.csv', sep=';')

print("Atualização concluída")
```
