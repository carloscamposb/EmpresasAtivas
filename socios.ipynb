##### 游릴游릴 Alterando Dtype para object 

```python
conversao = ['CNPJ BASICO']

# Converter as colunas espec칤ficas de 'int' para 'object'
dados[conversao] = dados[conversao].astype('object')

# Verificar os tipos de dados ap칩s a convers칚o
dados.dtypes
```

```python
# Definir as colunas espec칤ficas para verificar
colunas_data = ['DATA DE ENTRADA DA SOCIEDADE']
dados[colunas_data]
```

```python
# Substituir valores '0' ou inv치lidos por NaT (Not a Time)
dados['DATA DE ENTRADA DA SOCIEDADE'] = dados['DATA DE ENTRADA DA SOCIEDADE'].replace('0', pd.NaT)

# Converter para o formato datetime, for칞ando o tratamento de valores inv치lidos
dados['DATA DE ENTRADA DA SOCIEDADE'] = pd.to_datetime(dados['DATA DE ENTRADA DA SOCIEDADE'], errors='coerce', format='%Y%m%d')

# Formatar as datas para DD/MM/AAAA
dados['DATA DE ENTRADA DA SOCIEDADE'] = dados['DATA DE ENTRADA DA SOCIEDADE'].dt.strftime('%d/%m/%Y')

# Exibir as colunas para verificar a convers칚o
dados['DATA DE ENTRADA DA SOCIEDADE']
```
